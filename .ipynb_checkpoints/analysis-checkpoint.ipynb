```python
from model import PostConflictGrowthModel
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = PostConflictGrowthModel()

# Initial conditions
K0 = 100
H0 = 80
A0 = 1.0

T = 50  # time horizon
```

```python
K = np.zeros(T)
H = np.zeros(T)
A = np.zeros(T)
Y = np.zeros(T)

K[0] = K0
H[0] = H0
A[0] = A0

for t in range(T - 1):
    Y[t] = model.production(K[t], H[t], A[t])
    
    K[t + 1] = model.update_capital(K[t], Y[t])
    H[t + 1] = model.update_human_capital(H[t])
    A[t + 1] = model.update_governance(A[t])

Y[-1] = model.production(K[-1], H[-1], A[-1])
```

```python
plt.figure()
plt.plot(Y)
plt.title("Baseline GDP Path (No Conflict)")
plt.xlabel("Time")
plt.ylabel("Output (Y)")
plt.show()
```

```python
# --- Conflict Scenario ---

K_conf = np.zeros(T)
H_conf = np.zeros(T)
A_conf = np.zeros(T)
Y_conf = np.zeros(T)

K_conf[0] = K0
H_conf[0] = H0
A_conf[0] = A0

shock_time = 15

for t in range(T - 1):
    Y_conf[t] = model.production(K_conf[t], H_conf[t], A_conf[t])

    # Apply conflict shock once
    if t == shock_time:
        K_conf[t], H_conf[t], A_conf[t] = model.apply_conflict_shock(
            K_conf[t],
            H_conf[t],
            A_conf[t],
            capital_loss=0.4,
            human_capital_loss=0.3,
            governance_loss=0.2
        )

    K_conf[t + 1] = model.update_capital(K_conf[t], Y_conf[t])
    H_conf[t + 1] = model.update_human_capital(H_conf[t])
    A_conf[t + 1] = model.update_governance(A_conf[t])

Y_conf[-1] = model.production(K_conf[-1], H_conf[-1], A_conf[-1])
```

```python
plt.figure()
plt.plot(Y, label="Fragile Baseline")
plt.plot(Y_conf, label="Post-Conflict Economy")
plt.axvline(x=shock_time, linestyle='--')
plt.title("GDP Path: Fragile Economy vs Conflict Shock")
plt.xlabel("Time")
plt.ylabel("Output (Y)")
plt.legend()
plt.show()
```

```python
# --- Conflict + Capital Injection Only ---

K_cap = np.zeros(T)
H_cap = np.zeros(T)
A_cap = np.zeros(T)
Y_cap = np.zeros(T)

K_cap[0] = K0
H_cap[0] = H0
A_cap[0] = A0

for t in range(T - 1):
    Y_cap[t] = model.production(K_cap[t], H_cap[t], A_cap[t])

    # Conflict shock
    if t == shock_time:
        K_cap[t], H_cap[t], A_cap[t] = model.apply_conflict_shock(
            K_cap[t],
            H_cap[t],
            A_cap[t],
            capital_loss=0.4,
            human_capital_loss=0.3,
            governance_loss=0.2
        )

    # Capital injection after conflict
    if t == shock_time + 1:
        K_cap[t] += 50  # external reconstruction funds

    K_cap[t + 1] = model.update_capital(K_cap[t], Y_cap[t])
    H_cap[t + 1] = model.update_human_capital(H_cap[t])
    A_cap[t + 1] = model.update_governance(A_cap[t])

Y_cap[-1] = model.production(K_cap[-1], H_cap[-1], A_cap[-1])
```

```python
plt.figure()

plt.plot(Y, label="Fragile Baseline")
plt.plot(Y_conf, label="Conflict")
plt.plot(Y_cap, label="Conflict + Capital Injection")

plt.axvline(x=shock_time, linestyle='--')

plt.title("GDP Paths Under Different Post-Conflict Scenarios")
plt.xlabel("Time")
plt.ylabel("Output (Y)")
plt.legend()
plt.show()
```

```python
model = PostConflictGrowthModel()
sim = SimulationRunner(model, T=50)
```

```python
baseline = sim.run(100, 80, 1.0)

conflict = sim.run(
    100, 80, 1.0,
    shock_time=15,
    shock_params={
        "capital_loss": 0.4,
        "human_capital_loss": 0.3,
        "governance_loss": 0.2
    }
)

plot_gdp_paths({
    "Baseline": baseline,
    "Conflict": conflict
}, shock_time=15)
```

```python
conflict_capital = sim.run(
    100, 80, 1.0,
    shock_time=15,
    shock_params={
        "capital_loss": 0.4,
        "human_capital_loss": 0.3,
        "governance_loss": 0.2
    },
    capital_injection=50
)
```

```python
plot_gdp_paths({
    "Baseline": baseline,
    "Conflict": conflict,
    "Conflict + Capital Injection": conflict_capital
}, shock_time=15)
```

```python
conflict_human_reform = sim.run(
    100, 80, 1.0,
    shock_time=15,
    shock_params={
        "capital_loss": 0.4,
        "human_capital_loss": 0.3,
        "governance_loss": 0.2
    },
    policy_change={
        "education_investment": 0.06
    }
)
```

```python
plot_gdp_paths({
    "Baseline": baseline,
    "Conflict": conflict,
    "Capital Injection": conflict_capital,
    "Human Capital Reform": conflict_human_reform
}, shock_time=15)
```

```python
def find_education_threshold(sim_runner, K0, H0, A0, shock_time, shock_params):
    
    for edu in np.linspace(0.01, 0.15, 50):
        results = sim_runner.run(
            K0, H0, A0,
            shock_time=shock_time,
            shock_params=shock_params,
            policy_change={"education_investment": edu}
        )
        
        Y = results["Y"]
        
        # Check if final output is higher than mid-point (trend reversal proxy)
        if Y[-1] > Y[shock_time]:
            return edu
    
    return None
```

```python
threshold = find_education_threshold(
    sim,
    100, 80, 1.0,
    shock_time=15,
    shock_params={
        "capital_loss": 0.4,
        "human_capital_loss": 0.3,
        "governance_loss": 0.2
    }
)

threshold
```
